```python

def compare_devices(device_folders):
    # Initialize a list to store the plot objects for each device
    all_data = []  # Store dataframes for each device
    color_palette = ['b', 'g', 'r', 'c', 'm', 'y', 'k']  # Define color palette

    # Loop through each device folder
    for idx, device_folder in enumerate(device_folders):
        # Get color for this device
        color = color_palette[idx % len(color_palette)]

        # Call the inspect_device function for each device with the assigned color
        data = inspect_device(device_folder, color)
        all_data.append(data)

    # Concatenate data from all devices for each plot type
    combined_data = {
        'duration_histplot': [],
        'number_of_files_per_day': [],
        'number_of_files_per_hour_plot': [],
        'calendar_line_of_points_plot': [],
        'duration_vs_datetime_plot': []
    }

    for data in all_data:
        for i, key in enumerate(combined_data):
            if i == 0:
                combined_data[key] = data[i]
            else:
                combined_data[key] = combined_data[key].append(data[i])

    # Create combined plots with different colors for each device
    plt.figure(figsize=(10, 5))

    for idx, (key, data) in enumerate(combined_data.items()):
        plt.plot(data, label=f'Device {idx + 1}')

    plt.xlabel("X-axis Label")
    plt.ylabel("Y-axis Label")
    plt.legend()
    plt.title("Combined Plot")

    # Show the combined plot
    plt.show()
```

```python

# Example usage
main_folder = "/media/alvaro/BladeHDD/Bioacoustics_BigData/Monteverde_Ubuntutouch/Experimentos/Segundo/"
device_folders = [main_folder + "/Pixel3a/", main_folder + "/OneNote5/", main_folder + "/SamsungS7/"]
compare_devices(device_folders)
```
